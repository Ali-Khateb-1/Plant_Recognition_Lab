```python
import os

target_path = "/mnt/c/Users/ali/Desktop/intro-to-ml-tensorflow/projects/p2_image_classifier"

def analyze_structure(path, indent=0):
    try:
        for item in os.listdir(path):
            full_path = os.path.join(path, item)
            if os.path.isdir(full_path):
                print(" " * indent + item)
                analyze_structure(full_path, indent + 1)
            else:
                print(" " * indent + item)
    except FileNotFoundError:
        print("❌ المسار غير موجود:", path)

print(" تحليل هيكلية:", target_path)
analyze_structure(target_path)
```

```python
import sys
import yaml
import subprocess

env_data = {
    "python_version": sys.version.split()[0],
    "environment_path": sys.prefix,
}

def get_conda_packages():
    try:
        conda_list = subprocess.check_output(["conda", "list"]).decode().split("\n")
        return [line.split()[0] for line in conda_list if line and not line.startswith("#")]
    except Exception as e:
        return [str(e)]

def get_micromamba_packages():
    try:
        micromamba_list = subprocess.check_output(["micromamba", "list"]).decode().split("\n")
        return [line.split()[0] for line in micromamba_list if line and not line.startswith("#")]
    except Exception as e:
        return [str(e)]

def get_pip_packages():
    try:
        pip_list = subprocess.check_output(["pip", "list", "--format=freeze"]).decode().split("\n")
        return [line.split("==")[0] for line in pip_list if line]
    except Exception as e:
        return [str(e)]

env_data["installed_conda_packages"] = get_conda_packages()
env_data["installed_micromamba_packages"] = get_micromamba_packages()
env_data["installed_pip_packages"] = get_pip_packages()

yaml_path = "environment_status.yaml"
with open(yaml_path, "w") as file:
    yaml.dump(env_data, file, default_flow_style=False)

print(f"تم تحديث `{yaml_path}` بالمكتبات المثبتة يدوياً.")
```
