```python
import qrcode
from PIL import Image, ImageDraw
import os

# 🔹 إنشاء الباركود بالرابط المطلوب
qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
qr.add_data("https://ali-khateb-1.github.io/STEAM_Nexus_Hub/")
qr.make(fit=True)

# 🔹 إنشاء صورة الباركود الأساسية
qrcode_img = qr.make_image(fill="black", back_color="white").convert("RGBA")

# 🔹 تحديد حجم الباركود النهائي
size = qrcode_img.size
triangle_img = Image.new("RGBA", size, (255, 255, 255, 0))

# 🔹 رسم مثلث داخل الصورة
draw = ImageDraw.Draw(triangle_img)
draw.polygon([(0, size[1]), (size[0] // 2, 0), (size[0], size[1])], fill="black")

# 🔹 تطبيق شكل المثلث على الباركود عبر تقنية القص
mask = triangle_img.convert("L")
qrcode_img.putalpha(mask)

# 🔹 حفظ الصورة داخل المستودع
save_path = "/mnt/c/Users/ali/Desktop/STEAM_Nexus_Hub/Fast_Test/qrcode_triangle.png"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
qrcode_img.save(save_path)

print(f" تم حفظ الباركود بالشكل المثلث في: {save_path}")
```

```python
import qrcode
from PIL import Image, ImageDraw
import os

# 🔹 إعداد الرابط والصورة
url = "https://ali-khateb-1.github.io/STEAM_Nexus_Hub/"
image_path = "/mnt/c/Users/ali/Downloads/A clear, professiona.png"

# 🔹 إنشاء الباركود
qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
qr.add_data(url)
qr.make(fit=True)
qrcode_img = qr.make_image(fill="black", back_color="white").convert("RGBA")

# 🔹 تحميل الصورة وتكبيرها بنسبة 170%
logo = Image.open(image_path)
logo_size = (qrcode_img.size[0] // 3, qrcode_img.size[1] // 3)
logo = logo.resize(logo_size, Image.LANCZOS)

# 🔹 دمج الصورة داخل الباركود
pos = ((qrcode_img.size[0] - logo_size[0]) // 2, (qrcode_img.size[1] - logo_size[1]) // 2)
qrcode_img.paste(logo, pos, logo)

# 🔹 إنشاء خلفية تحتوي على 5 باركودات (الوسط + 4 إضافيات محيطة به)
canvas_size = (qrcode_img.size[0] * 3, qrcode_img.size[1] * 3)  # حجم الخلفية
canvas = Image.new("RGBA", canvas_size, (255, 255, 255, 0))

# 🔹 إضافة الباركودات الخمسة
positions = [
    (qrcode_img.size[0], qrcode_img.size[1]),  # الباركود الرئيسي في الوسط
    (0, qrcode_img.size[1]),  # يسار الوسط
    (qrcode_img.size[0] * 2, qrcode_img.size[1]),  # يمين الوسط
    (qrcode_img.size[0], 0),  # أعلى الوسط
    (qrcode_img.size[0], qrcode_img.size[1] * 2)  # أسفل الوسط
]

for pos in positions:
    canvas.paste(qrcode_img, pos, qrcode_img)

# 🔹 قص الصورة باستخدام دائرة تجعل الباركود المركزي كاملاً مع بقايا المحيطين به
mask = Image.new("L", canvas_size, 0)
draw = ImageDraw.Draw(mask)
circle_radius = qrcode_img.size[0] * 1.25  # قطر الدائرة أكبر بقليل من الباركود
circle_center = (canvas_size[0] // 2, canvas_size[1] // 2)
draw.ellipse(
    [circle_center[0] - circle_radius, circle_center[1] - circle_radius,
     circle_center[0] + circle_radius, circle_center[1] + circle_radius],
    fill=255)

canvas.putalpha(mask)

# 🔹 حفظ الصورة النهائية
save_path = "/mnt/c/Users/ali/Desktop/STEAM_Nexus_Hub/Fast_Test/qrcode_final.png"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
canvas.save(save_path)

print(f"✅ تم حفظ الباركود بالشكل المطلوب في: {save_path}")
```

```python
import os
import qrcode
from PIL import Image, ImageDraw

# 🔹 إعداد الرابط
url = "https://ali-khateb-1.github.io/STEAM_Nexus_Hub/"

# 🔹 تحديد اسم السكربت والمجلد
script_name = "mp1"
base_path = "/mnt/c/Users/ali/Desktop/STEAM_Nexus_Hub/Fast_Test/Masterpiece"
script_folder = os.path.join(base_path, script_name)
os.makedirs(script_folder, exist_ok=True)

# 🔹 إعداد الباركود العادي
qr = qrcode.QRCode(
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4
)
qr.add_data(url)
qr.make(fit=True)
qrcode_img = qr.make_image(fill="black", back_color="white").convert("RGBA")

# 🔹 تدوير الباركود بزاوية 45 درجة
rotated_qrcode = qrcode_img.rotate(45, expand=True)

# 🔹 إعداد حجم اللوحة لرسم المثلثات
canvas_size = rotated_qrcode.size
canvas = Image.new("RGBA", canvas_size, (255, 255, 255, 0))
draw = ImageDraw.Draw(canvas)

# 🔹 ضبط أبعاد المثلثات بحيث تطابق أبعاد الباركود المائل مع فراغ بسيط
triangle_size = qrcode_img.size[0] // 3
positions = [
    [(0, triangle_size), (triangle_size, 0), (triangle_size, triangle_size)],  # زاوية أعلى يسار
    [(canvas_size[0] - triangle_size, 0), (canvas_size[0], triangle_size), (canvas_size[0] - triangle_size, triangle_size)],  # زاوية أعلى يمين
    [(0, canvas_size[1] - triangle_size), (triangle_size, canvas_size[1]), (triangle_size, canvas_size[1] - triangle_size)],  # زاوية أسفل يسار
    [(canvas_size[0] - triangle_size, canvas_size[1]), (canvas_size[0], canvas_size[1] - triangle_size), (canvas_size[0] - triangle_size, canvas_size[1] - triangle_size)]  # زاوية أسفل يمين
]

# 🔹 قص الباركود داخل المثلثات
for pos in positions:
    draw.polygon(pos, fill="black")
    mask = Image.new("L", canvas_size, 0)
    mask_draw = ImageDraw.Draw(mask)
    mask_draw.polygon(pos, fill=255)
    qr_cropped = Image.composite(rotated_qrcode, canvas, mask)
    canvas.paste(qr_cropped, (0, 0), qr_cropped)

# 🔹 دمج الباركود المائل مع المثلثات
canvas.paste(rotated_qrcode, (0, 0), rotated_qrcode)

# 🔹 حفظ الناتج داخل المجلد الصحيح
output_path = os.path.join(script_folder, f"{script_name}.png")
canvas.save(output_path)

print(f"✅ تم إنشاء المجلد: {script_folder}")
print(f"✅ تم حفظ السكربت داخل المجلد.")
print(f"✅ تم حفظ الناتج النهائي في: {output_path}")
```
