```python
import qrcode
from PIL import Image, ImageDraw
import os

# 🔹 إعداد الرابط والصورة
url = "https://ali-khateb-1.github.io/STEAM_Nexus_Hub/"
image_path = "/mnt/c/Users/ali/Downloads/A clear, professiona.png"

# 🔹 إنشاء الباركود
qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
qr.add_data(url)
qr.make(fit=True)
qrcode_img = qr.make_image(fill="black", back_color="white").convert("RGBA")

# 🔹 تحميل الصورة وتكبيرها بنسبة 170%
logo = Image.open(image_path)
logo_size = (qrcode_img.size[0] // 3, qrcode_img.size[1] // 3)
logo = logo.resize(logo_size, Image.LANCZOS)

# 🔹 دمج الصورة داخل الباركود
pos = ((qrcode_img.size[0] - logo_size[0]) // 2, (qrcode_img.size[1] - logo_size[1]) // 2)
qrcode_img.paste(logo, pos, logo)

# 🔹 إنشاء خلفية تحتوي على 5 باركودات (الوسط + 4 إضافيات محيطة به)
canvas_size = (qrcode_img.size[0] * 3, qrcode_img.size[1] * 3)  # حجم الخلفية
canvas = Image.new("RGBA", canvas_size, (255, 255, 255, 0))

# 🔹 إضافة الباركودات الخمسة
positions = [
    (qrcode_img.size[0], qrcode_img.size[1]),  # الباركود الرئيسي في الوسط
    (0, qrcode_img.size[1]),  # يسار الوسط
    (qrcode_img.size[0] * 2, qrcode_img.size[1]),  # يمين الوسط
    (qrcode_img.size[0], 0),  # أعلى الوسط
    (qrcode_img.size[0], qrcode_img.size[1] * 2)  # أسفل الوسط
]

for pos in positions:
    canvas.paste(qrcode_img, pos, qrcode_img)

# 🔹 قص الصورة باستخدام دائرة تجعل الباركود المركزي كاملاً مع بقايا المحيطين به
mask = Image.new("L", canvas_size, 0)
draw = ImageDraw.Draw(mask)
circle_radius = qrcode_img.size[0] * 1.25  # قطر الدائرة أكبر بقليل من الباركود
circle_center = (canvas_size[0] // 2, canvas_size[1] // 2)
draw.ellipse(
    [circle_center[0] - circle_radius, circle_center[1] - circle_radius,
     circle_center[0] + circle_radius, circle_center[1] + circle_radius],
    fill=255)

canvas.putalpha(mask)

# 🔹 حفظ الصورة النهائية
save_path = "/mnt/c/Users/ali/Desktop/STEAM_Nexus_Hub/Fast_Test/qrcode_final.png"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
canvas.save(save_path)

print(f"✅ تم حفظ الباركود بالشكل المطلوب في: {save_path}")
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfds
import os

# 🔹 تحميل بيانات النباتات من TensorFlow Datasets
dataset_name = "tf_flowers"
(train_data, test_data), dataset_info = tfds.load(
    dataset_name, split=["train[:80%]", "train[80%:]"], as_supervised=True, with_info=True
)

# 🔹 تجهيز البيانات (تحجيم الصور وتطبيعها)
def preprocess_image(image, label):
    image = tf.image.resize(image, (224, 224)) / 255.0
    return image, label

train_data = train_data.map(preprocess_image).batch(32).shuffle(1000)
test_data = test_data.map(preprocess_image).batch(32)

# 🔹 تحميل نموذج MobileNet الأساسي من TensorFlow Hub
model = tf.keras.Sequential([
    hub.KerasLayer("https://tfhub.dev/google/imagenet/mobilenet_v2_100_224/classification/5", trainable=False),
    tf.keras.layers.Dense(dataset_info.features["label"].num_classes, activation="softmax")
])

# 🔹 تجميع النموذج
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# 🔹 تدريب النموذج
model.fit(train_data, epochs=5, validation_data=test_data)

# 🔹 حفظ النموذج المدرب داخل المسار الصحيح
output_folder = r"C:\Users\ali\Desktop\STEAM_Nexus_Hub\Classification_of_flower_structure_TM1\Plant_Recognition_Lab"
os.makedirs(output_folder, exist_ok=True)
model_save_path = os.path.join(output_folder, "trained_model.h5")
model.save(model_save_path)

print(f"✅ تم حفظ النموذج المدرب في: {model_save_path}")
```
